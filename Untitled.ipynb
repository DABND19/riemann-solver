```python
import json

import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = 'data/A05.json'

with open(filename) as f:
    data = json.load(f)
```

```python
x = data[0]['x']

density = [
    (item['t'], np.array(item['density']))
    for item in data
]
```

```python
T_START = 40
T_END = 48.58
```

```python
density = list(filter(lambda x: x[0] >= T_START and x[0] <= T_END, density))
```

```python
len(density)
```

```python
fig, ax = plt.subplots(figsize=(10, 8), dpi=200)

for time, rho in density[::15]:
    ax.plot(x, rho, label=f't={time}')

ax.set_xlabel('Radius')
ax.set_ylabel('Density')
ax.set_yscale('log')
ax.legend()
ax.grid(True)
ax.set_title(f'Amplitude=05')
```

```python
files = [
    'data/A05.json',
    'data/A33.json',
    'data/A66.json',
    'data/A98.json'
]

data = {}
for filename in files:
    _, _, A = filename.partition('data/A')
    A, _, _ = A.partition('.json')
    A = int(A)
    with open(filename) as f:
        data[A] = json.load(f)
```

```python
fig, ax = plt.subplots(figsize=(10, 8), dpi=200)

index = 100
for A, data_ in data.items():
    density_ = []
    t_ = []
    for layer in data_:
        t = layer['t']
        x = layer['x']
        density = layer['density']
        t_.append(t)
        density_.append(density[index])
    ax.plot(t_[600:1000], density_[600:1000], '-', label=f'A={A}%')

ax.set_xlabel('Time')
ax.set_ylabel('Density')
ax.set_title(f'Density R=1.35')
ax.legend()
ax.grid(True)
```
